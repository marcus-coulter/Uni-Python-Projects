# 2. Data Loading and Overview
Load the datasets and give a brief overview.

```python
# Load each CSV file into a pandas DataFrame
dividend_data = pd.read_csv('Companies_ranked_by_Dividend_Yield.csv')
earnings_data = pd.read_csv('Companies_ranked_by_Earnings.csv')
market_cap_data = pd.read_csv('Companies_ranked_by_Market_Cap.csv')
pe_ratio_data = pd.read_csv('Companies_ranked_by_P_E_ratio.csv')
revenue_data = pd.read_csv('Companies_ranked_by_Revenue.csv')

# Display the first few rows of each DataFrame
print(dividend_data.head())
print(earnings_data.head())
print(market_cap_data.head())
print(pe_ratio_data.head())
print(revenue_data.head())

# Check the shape and columns of each DataFrame
print("Dividend Data Columns:", dividend_data.columns)
print("Earnings Data Columns:", earnings_data.columns)
print("Market Cap Data Columns:", market_cap_data.columns)
print("P/E Ratio Data Columns:", pe_ratio_data.columns)
print("Revenue Data Columns:", revenue_data.columns)

```

## Merging Datasets


```python
# Rename columns in each dataset to avoid duplicates
dividend_data = dividend_data.rename(columns={'Rank': 'Dividend_Rank', 'country': 'Dividend_Country', 'price (GBP)': 'Dividend_Price'})
earnings_data = earnings_data.rename(columns={'Rank': 'Earnings_Rank', 'country': 'Earnings_Country', 'price (GBP)': 'Earnings_Price'})
market_cap_data = market_cap_data.rename(columns={'Rank': 'MarketCap_Rank', 'country': 'MarketCap_Country', 'price (GBP)': 'MarketCap_Price'})
pe_ratio_data = pe_ratio_data.rename(columns={'Rank': 'PERatio_Rank', 'country': 'PERatio_Country', 'price (GBP)': 'PERatio_Price'})
revenue_data = revenue_data.rename(columns={'Rank': 'Revenue_Rank', 'country': 'Revenue_Country', 'price (GBP)': 'Revenue_Price'})

# Now merge datasets on 'Name' and 'Symbol'
merged_data = (dividend_data
                .merge(earnings_data, on=['Name', 'Symbol'])
                .merge(market_cap_data, on=['Name', 'Symbol'])
                .merge(pe_ratio_data, on=['Name', 'Symbol'])
                .merge(revenue_data, on=['Name', 'Symbol']))

# Display the merged data
print(merged_data.head())
```
